AGE DISTRIBUTION

```python
df['age'].sort_values().iplot(kind='box', title='AGE DISTRIBUTION', yTitle = 'Age', theme='solar', color='cyan')
```

There exists a character with '?' in age

```python
df[df['age'] == '?']
```

2 rows have to cleaned for '?' values

```python
df.loc[df['age'] == '?', ['age']] = float('nan')
df.dropna(inplace=True)
```

```python
df['age'] = df['age'].astype('int')
df['age'].dtype
```

Now Visualising Age Distribution again

```python
df['age'].sort_values().iplot(kind='box', title='AGE DISTRIBUTION', yTitle = 'Age', theme='solar', color='cyan', boxpoints="outliers")
```

There exists a row with age = 383 and other outliers with age > 59. Removing 383 

```python
df.drop(df[df['age'] == 383].index, inplace=True)
```

Visualising again

```python
df['age'].sort_values().iplot(kind='box', title='AGE DISTRIBUTION', yTitle = 'Age', theme='solar', color='cyan', boxpoints="outliers")
```

Visualising each score count distribution between 0 and 1

```python
df.astype('string').groupby('A2_Score').count()['age'].iplot(kind='bar',title=' Distribution', yTitle='count', theme='solar', color='cyan')
```

```python
for i in range(1, 11):
    name = 'A' + str(i) + '_Score'
    print(name)
    df.astype('string').groupby(name).count()['age'].iplot(kind='bar',title=name + ' Distribution', yTitle='count', theme='solar', color='cyan', xTitle=name)
    # fig.show()
```

A1, A6, A7, A9, A10 have skewed responses while others have equal number of responses

Visualising each score distribution with class

```python
for i in range(1, 11):
    name = 'A' + str(i) + '_Score'
    fig = px.histogram(df, x=name, color='pred', barmode='group', template='plotly_dark', color_discrete_sequence=px.colors.qualitative.Prism)
    fig.show()
```

**Inference:**  
A1_Score: Highly Skewed. Most people with autism have chosen 1. People without autism also have selected 1  
  
A2_Score: Most people with autism have chosen 1, but has equal number of people without autism choosing this option  
  
A3_Score: Most people with autism have chosen 1, but has equal number of people without autism choosing this option 
   
A4_Score: Most people with autism have chosen 1, but has equal number of people without autism choosing this option   
  
A5_Score: Highly Skewed. Most people with autism have chosen 1. No of autism people with 1 option is higher than people without autism.  
  
A6_Score: Distinct Difference. People with autism have chosen 1 and without have chosen 0 in major cases    
  
A7_Score: Most people with autism have chosen 1, but has equal number of people without autism choosing this option   
  
A8_Score: Highly Skewed. Most people with autism have chosen 1. People without autism also have selected 1  
  
A9_Score: Distinct Difference. People with autism have chosen 1 and without have chosen 0 in major cases  
  
A10_Score: People with autism have chosen 1 in most cases but people without autism have chosen both options almost equally
